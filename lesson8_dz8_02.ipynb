```python
from scipy import stats
from scipy.stats.mstats import winsorize
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('cleaned_googleplaystore.csv')
```

```python
z_scores = stats.zscore(df['Installs'])
z_scores
```

```python
threshold = 3
outliers = df['Installs'][z_scores > threshold]
df.loc[z_scores > threshold, 'Installs'] = df['Installs'].median()
```

```python
plt.hist(df['Installs'])
plt.xlabel('Installs')
plt.ylabel('Count')
plt.show()
```

```python
z_scores = stats.zscore(df['Installs'])
threshold = 3
outliers = df['Installs'][z_scores > threshold]
df.loc[z_scores > threshold, 'Installs'] = df['Installs'].median()
```

```python
df['Rating'] = winsorize(df['Rating'], limits=[0.05, 0.05])
```

```python
plt.hist(df['Rating'])
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
scaler = StandardScaler()
df[['Rating']] = scaler.fit_transform(df[['Rating']])
```

```python
plt.hist(df['Rating'])
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
